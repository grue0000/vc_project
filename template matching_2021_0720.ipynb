```python
# 동영상에서 골프공 인식

# 시간 : t = 1/fps   x속도 : vx = (x-lx)/t    , y속도 : vy = (y-ly)/t

import cv2
import numpy as np

cap = cv2.VideoCapture('vid.mp4')

templit = cv2.imread("ball.jpg", cv2.IMREAD_GRAYSCALE)

width = int(cap.get(3)) # 가로 길이 가져오기 
height = int(cap.get(4)) # 세로 길이 가져오기
fps = 20

fcc = cv2.VideoWriter_fourcc('D', 'I', 'V', 'X')  # 윈도우 코덱
out = cv2.VideoWriter('output.avi', fcc, fps, (width, height))

t = 1/fps
lx = 0
ly = 0

while (cap.isOpened()) :
    ret, frame = cap.read()
    if ret :
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        
        result = cv2.matchTemplate(gray, templit, cv2.TM_CCORR_NORMED) # 동영상은 TM_CCORR_NORMED이 더 좋음

        minVal, maxVal, minLoc, maxLoc = cv2.minMaxLoc(result)
        x, y = minLoc
        h, w = templit.shape
        
        gray = cv2.rectangle(frame, (x, y), (x +  w, y + h) , (0, 0, 255), 1)
        out.write(frame)
        cv2.imshow('frame', frame)
        
        vx = (x-lx)/t
        vy = (y-ly)/t
        
        print('x속도 : ',vx,'y속도 : ',vy)
        
        lx, ly = x, y
        
        if cv2.waitKey(1) & 0xFF == ord('q') : break
    else :
        print("Fail to read frame!")
        break

cap.release()
out.release()
cv2.destroyAllWindows()
```
